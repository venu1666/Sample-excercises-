```python
# importing libs
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#loading dataset
dataset=pd.read_csv('alcoho_available.csv')
dataset.isnull().sum()
```

```python
dataset=dataset.drop(columns=['Series_title_2','Series_title_3','Series_title_4','Series_title_5'])
```

```python

dataset.dtypes
```

```python
dataset.Data_value.fillna(dataset.Data_value.mean(),inplace=True)
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
L=LabelEncoder()
dataset=dataset.apply(L.fit_transform)

dataset
```

```python
dataset=pd.get_dummies(dataset,columns=['Series_reference'])
```

```python
dataset
```

```python
y=dataset.pop('Data_value')
X=dataset
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
L=LinearRegression()
L.fit(X_train,Y_train)
```

```python
pred=L.predict(X_test)
pred
```

```python
from sklearn.metrics import explained_variance_score
explained_variance_score(Y_test,pred)
```
