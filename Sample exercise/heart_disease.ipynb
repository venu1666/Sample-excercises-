```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('processed.cleveland.data',na_values=['?'])
dataset
```

```python
dataset.columns=["age", 
"sex", 
"cp", 
"trestbps", 
"chol", 
"fbs", 
"restecg", 
"thalach", 
"exang", 
"oldpeak", 
"slope", 
"ca", 
"thal", 
"num (the predicted attribute)" ]
```

```python

dataset.sex.fillna(dataset.sex.mode().values[0],inplace=True)
```

```python
dataset.cp.fillna(dataset.cp.mode().values[0],inplace=True)
```

```python
import matplotlib.pyplot as plt
plt.subplot(3,3,1)
dataset.trestbps.hist()
plt.subplot(3,3,2)
dataset.chol.hist()
plt.subplot(3,3,3)
dataset.fbs.hist()
plt.subplot(3,3,4)
dataset.restecg.hist()
plt.subplot(3,3,5)
dataset.thalach.hist()
```

```python
dataset.trestbps.fillna(dataset.trestbps.median(),inplace=True)
dataset.chol.fillna(dataset.chol.median(),inplace=True)
dataset.fbs.fillna(dataset.fbs.mode().values[0],inplace=True)
dataset.restecg.fillna(dataset.restecg.mode().values[0],inplace=True)
dataset.thalach.fillna(dataset.thalach.median(),inplace=True)
dataset.thal.fillna(dataset.thal.median(),inplace=True)
dataset.ca.fillna(dataset.ca.median(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
import seaborn as sns
sns.pairplot(dataset,hue="num (the predicted attribute)")
```

```python

y=dataset.pop('num (the predicted attribute)')
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(dataset,y,test_size=0.1,random_state=0)
```

```python
X_train.shape,Y_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
D=DecisionTreeClassifier()
D.fit(X_train,Y_train)
pred=D.predict(X_test)

#help(GridSearchCV)
from sklearn.model_selection import GridSearchCV
p = {'min_samples_leaf':range(10,200,10),'min_samples_split':range(10,200,10),'max_depth':range(10,200,10)}
s=GridSearchCV(D,p,cv=2)
s.fit(X_train,Y_train)
s.best_estimator_
```

```python
p={'min_samples_leaf':[10],'min_samples_split':[10],'max_depth':[10]}
s=GridSearchCV(D,p,cv=5)
s.fit(X_train,Y_train)
pred=s.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,pred)
```
