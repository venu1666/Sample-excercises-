```python
#import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.cross_validation import train_test_split
```

```python
dataset=dataset.apply(LabelEncoder().fit_transform)
dataset.head()
```

```python
print(dataset.Type.value_counts())
print(" ")
print(dataset.Units.value_counts())
print(" ")
print(dataset.Indicator.value_counts())
print(" ")
print(dataset.Cause.value_counts())
print(" ")
print(dataset.Validation.value_counts())
print(" ")
print(dataset.Population.value_counts())
print(" ")
print(dataset.Age.value_counts())
print(" ")
```

```python
#sns.pairplot(dataset,hue='Severity')
y=dataset.pop('Severity')
X=dataset
```

```python
X=pd.get_dummies(dataset,columns=['Indicator'])
```

```python
X.shape
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
D=DecisionTreeClassifier()
D.fit(X_train,Y_train)
pred=D.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,pred)
```
